### Gradient descent disadvantage! 

consider the following function: $$f(x)=x^4 - 3x^3 +2$$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x):
    return x**4 - 5*x**3 + 2
```

```python
x1 = np.linspace(start=-3, stop=6, num=500)
```

```python
plt.xlim(-3, 6)
#plt.ylim(0, 8)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.plot(x1, f(x1))
plt.show()
```

```python
x0    = -4 # initial value
epoch = 100
alpha = 0.0001
```

```python
def gradient(x):
    return 4*x**3 - 15*x**2 
```

```python
x_star = []
for i in range(epoch):
    x_star.append(x0 - alpha * gradient(x0))
    x0 = x_star[-1]
```

```python
#x_star
```

```python
plt.plot(x_star)
plt.show()
```

```python
min_f = f(x_star[-1])
```

```python
print(f'The minimum f(x) is {min_f} which happens at x={x_star[-1]}'  )
```

```python
plt.xlim(-3, 6)
#plt.ylim(0, 8)
plt.xlabel('X', fontsize=16)
plt.ylabel('f(x)', fontsize=16)
plt.plot(x1, f(x1))
plt.axvline(x=x_star[-1], lw=1, color='red', ls = '--')
plt.axhline(y=f(x_star[-1]), lw=1, color='red', ls = '--')


plt.show()
```
